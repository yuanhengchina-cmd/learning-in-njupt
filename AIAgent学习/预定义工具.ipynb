```python
from langchain_tavily import TavilySearch
#初始化工具,现在就是一个工具了，可以直接调用它，也可以把他传给智能体
search_tool = TavilySearch(#官方提供好的用来约束搜索行为的参数
    max_results=5,#控制搜索的数据条数（找出关联度最高的几个网页)
    topic="general",#搜索时偏向于哪一类数据，news新闻相关的，general通用的，finance金融相关的（3种）
    # include_answer=False,
    # include_raw_content=False,
    # include_images=False,
    # include_image_descriptions=False,
    # search_depth="basic",
    # time_range="day",
    # include_domains=None,
    # exclude_domains=None
)
```

```python
search_tool.invoke("胖猫是什么梗？")
```

 'results'是搜索的结果，搜索的不止一条，是把网页的信息提取出来返回，是一个数组，不止一条；有'url'、  'score'

```python
from langchain.agents import create_agent
#创建智能体，使用预定义工具tavily
agent = create_agent(
    model="deepseek-chat",
    tools=[search_tool],
    system_prompt="你是一个智能助手，你使用工具来解决用户的问题"
)#如果我们没提供工具给他，他的结果一定是根据用户数据回答的，而不是查找的结果
```

```python
from langchain.messages import HumanMessage
response=agent.invoke(
    {"messages":[HumanMessage("胖猫是什么梗？")]}
)
for message in response["messages"]:
    message.pretty_print()
#速度慢的原因在于ai可能经过多次搜索
```

使得智能体能够接触网络与网络对接，基于最新实时发生的消息回答，而不是仅基于自己训练的数据

```python
#优化/问题
#token消耗，我们使用的官方的search_tool = TavilySearch实现/定义时特别复杂，内部有很多参数，后果是消耗更多的token；当我们任务比较简单，有时只要一个Query条件就行。
#第二个问题，ai回答时的丰富度还行，但是可信度不是很高，有ai幻觉，缺少罗列回答的来源和根据（类似于论文的引用、来源标记）
from langchain.tools import tool
"""
解决思路
-自定义tavily工具（简化参数）
-结构化输出
"""
#先使用官方的客户端初始化/也可以使用tavily官方的SDK，tavily_client也行
tavily=TavilySearch(max_results=5,topic="general",)
#我们不直接传给智能体，而是先封装一下
#然后自己封装tool,里面还是使用的langchain官方的工具
#我们传递给模型的信息就是当前工具定义的信息，一定程度上简化
@tool
def web_search(query:str):
    """
   search the web for information
    """
    return  tavily.invoke(query)
```

```python
from typing import List
#解决可信度低，不包含网页信息源
#使用结构化输出来解决
from pydantic import BaseModel,Field
#Agent回答内容引用的网页信息
class Reference(BaseModel):
    title:str=Field(description="The tile of the web page cited in the answer")
    url: str=Field(description="The url of the web page cited in the answer")
#Agent的回答内容,引用是列表，（可能引用到的网页都传给我）将多个网页传给我（每个网页都要传具体的网页和标题），以便我们后续在页面上渲染出来，以便我们可以看具体引用可以点击看后续
class AnswerInfo(BaseModel):
    answer:str=Field(description="The final answer for user")
    reference:List[Reference]=Field(description="The web page cited in the answer")
```

```python
agent=create_agent(
    model="deepseek-chat",
    tools=[web_search],
    system_prompt="你是一个智能助手，你是用工具来解决用户问题",
    response_format=AnswerInfo
)
```

```python
from langchain.messages import HumanMessage
response=agent.invoke(
    {"messages":[HumanMessage("胖猫是什么梗？")]}
)
print(response)
```
